```python
import requests
# Create an URL object
url = 'https://stats.espncricinfo.com/ci/engine/records/team/match_results.html?id=14450;type=tournament'
# Create object page
page = requests.get(url)
print(page)
# If the output is <Response [200]> that means the server allows us to collect data from their website.
```

```python
from bs4 import BeautifulSoup
# parser-lxml = Change html to Python friendly format
# Obtain page's information
soup = BeautifulSoup(page.text, 'lxml')
```

```python
matches_table = soup.select("table.engineTable")[0]
```

```python
headers = []
for i in matches_table.find_all('th'):
  title = i.text
  headers.append(title)
print(headers)
```

```python
import pandas as pd
matches_data = pd.DataFrame(columns = headers)
```

```python
# Create a for loop to fill mydata
for j in matches_table.find_all('tr')[1:]:
  row_data = j.find_all('td')
  row = [i.text for i in row_data]
  length = len(matches_data)
  matches_data.loc[length] = row
matches_data
```

```python
# Export to csv
matches_data.to_csv('WC_matches_22.csv', index=False)
```
